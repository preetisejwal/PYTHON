```python
df_bank=pd.read_csv('Downloads/Bank Personal Loan Modelling.csv')
```

```python
df_bank
```

```python
summary = df_bank.describe(include='all')
print(summary)
```

```python
# Mean for all numeric columns
mean_values = df_bank.mean()

# Median for all numeric columns
median_values = df_bank.median()

# Mode for all numeric columns (returns the most frequent value)
mode_values = df_bank.mode()
```

```python
# Range for all numeric columns (max - min)
range_values = df_bank.max() - df_bank.min()

# Variance for all numeric columns
variance_values = df_bank.var()

# Standard deviation for all numeric columns
std_dev_values = df_bank.std()

# Interquartile range (Q3 - Q1)
IQR_values = df_bank.quantile(0.75) - df_bank.quantile(0.25)
```

```python
# Display all results
print("Mean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMode Values:\n", mode_values)
print("\nRange Values:\n", range_values)
print("\nVariance Values:\n", variance_values)
print("\nStandard Deviation Values:\n", std_dev_values)
print("\nInterquartile Range (IQR) Values:\n", IQR_values)
```

12. What is the percentage of variation you can observe in the ‘Income’ variable?

```python
#A:12  Calculate mean and standard deviation for the 'Income' variable
income_mean = df_bank['Income'].mean()
income_std_dev = df_bank['Income'].std()

# Calculate Coefficient of Variation (CV) as a percentage
cv = (income_std_dev / income_mean) * 100

# Display the result
print(f"The percentage of variation in 'Income' is:", cv)
```
